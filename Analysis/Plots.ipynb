```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
pval_data = pd.read_csv('pvaluesSentDiffSignifance.csv')
```

```python
ps = pval_data['x'].dropna()
```

```python
plt.figure(figsize=(14,14))
plt.plot(ps, 'o-')
plt.axhline(y=0.05)
plt.title("p-value for Sentiment Significance")
plt.xlabel('Game Time')
plt.ylabel('p-value')
plt.legend(['p-value', '0.05 Level'])
plt.savefig('pValuePlot.pdf')
```

```python
residualdf = pd.read_csv('Residuals.csv')
```

```python
residualdf.head()
```

```python
sent = residualdf['SentResid']
std = residualdf['StandardResid']
```

```python
plt.figure()
plt.plot(sent, 'o-')
plt.plot(std, 'b^-')
plt.legend(['Sentiment','Standard'], loc=2)
plt.title('Residuals for Win Probability Models')
plt.xlabel('Game Minute')
plt.ylabel('Residual')
plt.savefig('ResidualPlot.pdf')
```

```python
residualdf = pd.read_csv('VolumeSentResiduals.csv')
sent = residualdf['sent']
std = residualdf['standards']
vol = residualdf['volume']
plt.figure(figsize=(14,14))
plt.plot(sent, 'o-')
plt.plot(vol, 's--')
plt.plot(std, 'b^-')
plt.legend(['sentiment', 'volume','standard'], loc=2)
plt.title('Residuals for Win Probability Models')
plt.xlabel('Game Minute')
plt.ylabel('Residual')
plt.savefig('VolumeSentResidPlot.pdf')
```

```python
residualdf = pd.read_csv('CombinedResiduals.csv')
sent = residualdf['sent']
std = residualdf['standards']
vol = residualdf['volume']
combined = residualdf['combined']
plt.figure(figsize=(14,14))
plt.plot(sent, 'o-')
plt.plot(vol, 's--')
plt.plot(std, 'b^-')
plt.plot(combined, 'kh-')
plt.legend(['sentiment', 'volume','standard'], loc=2)
plt.title('Residuals for Win Probability Models')
plt.xlabel('Game Minute')
plt.ylabel('Residual')
plt.savefig('CombinedResidPlot.pdf')
```
